### The nomenclature followed in the below function is 

if the actual and predicted lael matches, then the score would be RA

if actual value exists but the predicted value is null, which means though i have value in form but bounding box not able to find that,
then the score would be WA1

if the actual value doesn't exists, but the bounding box just appears , which means he bounding box is not expected but still it appears
then the score would be WA2


Another very unlikely case would be that, if the bounding box predicted value is completely different from the actual value,
then the score would be WA




```python
def mycalc(table_name, predColName, actualColName1):
    
    
    
    if table_name[predColName] == -1 and table_name[actualColName] == -1:
        return "INV"
    elif table_name[predColName]==table_name[actualColName]:
        return "RA"
    elif table_name[predColName] == -1:
        return "WA1"
    elif table_name[actualColName] == -1:
        return "WA2"
    
    else:
        return "WA"
```

```python
performance_metrics={}

for col in score_cols_list:
    performance_metrics[col]={}
    print("my datatype",merged_results[col].dtype)
    
    
    tempStr=col
    score_string_removed=tempStr.replace('score_', '')
#     print("score string removed", score_string_removed)
    actual_string_added=score_string_removed+'_actual'


    replace_for_nan(merged_results, score_string_removed)
    replace_for_nan(merged_results, actual_string_added)

    
    merged_results[col] = merged_results.apply( lambda row : mycalc(row, score_string_removed, actual_string_added), axis = 1)
    
    
#     print(f"the no of RA's for {col} is {len(merged_results[merged_results[col]=='RA'])}")
#     print(f"the no of non zero predicted values's for {col} is {len(merged_results[merged_results[score_string_removed] != -1])}")
#     print(f"the no of non zero actual values's for {col} is {len(merged_results[merged_results[actual_string_added]!= -1])}")

    performance_metrics[col]['Precision']=len(merged_results[merged_results[col]=='RA'])/(len(merged_results[merged_results[score_string_removed] != -1]))
    performance_metrics[col]['Recall']=len(merged_results[merged_results[col]=='RA'])/(len(merged_results[merged_results[actual_string_added] != -1]))
    
#     print("MRGING..",merged_results[score_string_removed])
#     if merged_results[score_string_removed]==merged_results[expected_string_added]:
#         merged_results[cols]=1

# merged_results       
```

In order to calculate precision and recall, we are taking the fieldwise metrics.

**Precision** refers to out of all the values predicted (say Insured ABN), how many the model get it right

Precision= number of RA's for a particular Field/No of non zero predicted values for the particular field 



**Recall** refers to out of all actual ABN labels, how many the model get it right

Recall= number of RA's for a particular Field/No of non zero actual values for the particular field

```python
def assign_score(row):
    if row['predicted_value'] == row['expected_value']:
        return 'RA'
    elif (row['expected_value']==-1) and (row['predicted_value'] != -1):
        return 'WA2'
    elif (row['predicted_value']==-1) and (row['expected_value']!=-1):
        return 'WA1'
    else:
        return 'WA'
```

```python
df['score'] = df.apply(lambda row: assign_score(row), axis=1)
```

```python
df
```

```python
df = df.groupby('field')['score'].value_counts().unstack(fill_value=0).reset_index()
```

```python
df
```

```python
df["total_doc"]=df['RA']+df['WA']+df['WA1']+df['WA2']
```

```python
df
```

```python
df['accuracy']=df["RA"]/df["total_doc"]
```

```python
df['precision']=df["RA"]/(df["total_doc"]-df["WA1"])
```

```python
df['recall']=df["RA"]/(df["total_doc"]-df["WA2"])
```
